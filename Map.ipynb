```python
#importing libraries

import pandas as pd
import folium
import arrow
from collections import namedtuple
import numpy as np
import math 
```

```python
#load a folium world map
                
world = folium.Map(
    zoom_start=2,
    location=[13.133932434766733, 16.103938729508073])
```

```python
#load CSV file with all the latitudes and longtiudes

cities = pd.read_csv('cities.csv')
print(cities.shape)
cities.head(3)
```

```python
#add markers to map


world_all_cities= folium.Map(
    zoom_start=2,
    location = [13.133932434766733, 16.103938729508073]
)

for _, city in cities.iterrows():
    folium.Marker(
        location=[city['Latitude'], city['Longitude']],
    ).add_to(world_all_cities)
```

```python
#add popup with names to world map

world_all_cities_popup = folium.Map(
    zoom_start=2,
    location=[13.133932434766733, 16.103938729508073]
)

for _, city in cities.iterrows():
    folium.Marker(
        location=[city['Latitude'], city['Longitude']],
        popup=city['Ports'],
    ).add_to(world_all_cities_popup)
    
world_all_cities_popup
```

```python
world_all_cities_popup.save("KAMSARMAXMAP.html")
```
